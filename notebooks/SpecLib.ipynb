```python
%load_ext autoreload
%autoreload 2
import sys
sys.path.append('../code')
from spec_lib import SpecLib
sys.path.append('/home/simon/git/pymzmine/')
```

```python
# sl = SpecLib('../lib/ALL_GNPS.mgf')
sl = SpecLib('/srv/data/mibig-links/20190805/matched_mibig_gnps_update.mgf')
sl._load_mgf(id_field = 'SPECTRUMID')
```

```python
sl.filter()
```

```python
import jsonpickle
frozen = jsonpickle.encode(sl)
```

```python
with open('../lib/matched_mibig_gnps.p','w') as f:
    f.write(frozen)
```
